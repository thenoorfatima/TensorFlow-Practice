```python
from PIL import Image
import os
```

```python
root = ""
save = ""
```

```python
files = []
# r=root, d=directories, f = files
for r, d, f in os.walk(root):
    for file in f:
        files.append(file)
```

```python
i = 1
for image in files:
    im = Image.open(root + image)
    im.save(save + str(i) + '.PNG')
    i += 1
```
